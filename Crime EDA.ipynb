```python
#Dropping unnecessary columns
df.drop(['Longitude', 'Latitude', 'Unnamed: 0', 'ID' ], axis=1)

df = df.drop(['Longitude', 'Latitude', 'Unnamed: 0', 'ID' ], axis=1)

df.head()
```

```python
#convert string to datetime

df['Date'] = df['Date'].apply(pd.to_datetime)

df.head()
```

```python
def extract_month(Date):
    return Date.month

df['month'] = df['Date'].apply(extract_month)

df.head()
```

```python
def extract_day(Date):
    return Date.day

df['day'] = df['Date'].apply(extract_day)

df.head()
```

```python
def determine_season(month):
    if month in [3, 4, 5]:
        return 'Spring'
    elif month in [6, 7, 8]:
        return 'Summer'
    elif month in [9, 10, 11]:
        return 'Autumn'
    else:
        return 'Winter'
    
df['season'] = (df['Date'].dt.month % 12 + 3) // 3 # 1: Winter, 2: Spring, 3: Summer, 4: Fall

df.head()
```
